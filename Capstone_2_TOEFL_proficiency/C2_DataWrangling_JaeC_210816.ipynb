```python
# Read sheet names of excel file of interest
xl = pd.ExcelFile('DataSet/ICNALE_Infosheet_201912.xlsx')
xl.sheet_names
```

```python
# Load/read sheets of interest: WE_file, EE_file
WE_file = pd.read_excel('ICNALE_Infosheet_201912.xlsx', sheet_name = 'Written Essays_2.4')
EE_file = pd.read_excel('ICNALE_Infosheet_201912.xlsx', sheet_name = 'Edited Essays_2.1')
```

```python
EE_file.dtypes
```

```python
# Clean up column names - they're currently in the second row
EE_file.columns = EE_file.iloc[0]
EE_file = EE_file.drop(0)
```

```python
EE_file = EE_file.reset_index()
```

```python
EE_file.head()
```

```python
# Define and save the columns of interest - only the ID Code and the scores are of interest from EE_file: EE_scores
score_cols = ['Content (/12)', 'Organization (/12)', 'Vocabulary (/12)', 
              'Language Use (/12)', 'Mechanics (/12)', 'Total 1 (%)', 'Total 2 (Weighted %)']
EE_scores = EE_file[['File Name'] + score_cols].copy()
```

```python
EE_scores.head()
```

```python
# Rename ID to same identifier as WE_file
EE_scores = EE_scores.rename(columns = {'File Name' : 'Code'})
```

```python
# Format Code to be of same format as WE_file
for i, user in enumerate(EE_scores['Code']):
    EE_scores['Code'][i] = user[:6] + user[11:14]
```

```python
EE_scores.head()
```

```python
# How many unique Codes? Some are duplicates
EE_scores['Code'].nunique()
```

```python
# Convert scores to floats
EE_scores[score_cols] = EE_scores[score_cols].astype('float64', errors = 'raise')
```

```python
EE_scores.isna().sum()
```

```python
EE_scores.dtypes
```

```python
# Tidy EE_scores by averaging each individual's (Code's) essay scores
EE_unique = EE_scores.groupby('Code')[score_cols].mean()
```

```python
# Ready to merge
EE_unique.head()
```

```python
# Join EE_unique with WE_file. Left join with EE_unique - all entries without essay scores are meaningless
df = EE_unique.join(WE_file.set_index('Code'), how='left')
```

```python
df.shape
```

```python
df.dtypes
```

```python
# The following columns are the wrong format. Coerce to numeric type.
num_cols = ['Score', 'NS', 'Pronunciation', 'Presentation', 'EssayW']
```

```python
for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors = 'coerce')
```

```python
df.dtypes
```

```python
df = df.reset_index()
df.head()
```

```python
# View sample entry
df.iloc[0].T
```

```python
df.isna().sum()
```

```python
# Following columns are completely empty. Drop them.
df = df.drop(['PTJ2 (wds)', 'SMK2 (wds)', 'Self Ev', 'Yrs of Stay (< Yrs)', 'ENS Type'], axis=1).copy()
```

```python
df.isna().sum()
```

```python
df['Sex'].str.upper()
```

```python
df['Grade/Degree'].unique()
```

```python
# Fill NaN Grade/Degrees to zero, and create a column to indicate their .isna()
df['Grade_isna'] = df['Grade/Degree'].isna()
df['Grade/Degree'] = df['Grade/Degree'].fillna(0)
```

```python
df['Major/ Occupation'].unique()
```

```python
# Fill NaN Major/ Occupation to 'N/A', and create a column to indicate their .isna()
df['Major_isna'] = df['Major/ Occupation'].isna()
df['Major/ Occupation'] = df['Major/ Occupation'].fillna('N/A')
```

```python
df['Acad. Genre'].unique()
```

```python
# Fill NaN Acad. Genre to 'N/A', and create a column to indicate their .isna()
df['Acad_isna'] = df['Acad. Genre'].isna()
df['Acad. Genre'] = df['Acad. Genre'].fillna('N/A')
```

```python
df['Test'].unique()
```

```python
# Fill NaN Tests to 'N/A', and create a column to indicate their .isna()
df['Test_isna'] = df['Test'].isna()
df['Test'] = df['Test'].fillna('N/A')
```

```python
df['Score'].unique()
```

```python
# Fill NaN Scores to zero, and create a column to indicate their .isna()
df['Score_isna'] = df['Score'].isna()
df['Score'] = df['Score'].fillna(0)
```

```python
mask = df['Score'] == 0
mask
```

```python
# It's meaningless to have values in 'Test' without corresponding values in 'Score'. Overwrite them to 'N/A'.
series = df['Test'].copy()
series[mask] = 'N/A'
df['Test'] = series
(df['Test']  == 'N/A').sum()
```

```python
(df['Score'] == 0).sum()
```

```python
# Check if the entry missing INTM is missing all the other single missing values as well
df[df['INTM'].isna()].T
```

```python
# Fill the missing entries with median values instead of zeros and creating extra columns for all
missing_cols = ['INTM', 'INSM', 'INTM+INSM', 'INTM-INSM', 'Primary', 'Secondary', 'College', 'Inschool', 'Outschool', 
                'Listening', 'Reading', 'Speaking', 'Writing', 'NS', 'Pronunciation', 'Presentation', 'EssayW']
```

```python
for c in missing_cols:
    df[c] = df[c].fillna(df[c].median())
```

```python
# No more .isna() - double check
df.isna().sum()
```

```python
# For expanding the 'Test' row, create a copy: df_tests
df_tests = df.copy()
```

```python
# list of all unique test in Test: tests
tests = df['Test'].unique()
```

```python
# Initailize the new columns in df_tests
for test in tests:
    df_tests[test] = 0
```

```python
# Subset scores to their own columns
for i, group in df_tests.groupby('Test'):
    df_tests[str(i)] = group['Score']
    df_tests[str(i)] = df_tests[str(i)].fillna(0)
```

```python
df_tests.head(50)
```

```python
# Drop the Test and Score columns - they're now duplicate information
df_tests.drop(['Test', 'Score'], axis = 1)
```
