```python
# Import modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
pd.options.display.max_columns = 1000
```

```python
# Read sheet names of excel file of interest
xl = pd.ExcelFile('DataSet/ICNALE_Infosheet_201912.xlsx')
xl.sheet_names
```

```python
# Load/read sheets of interest: WE_file, EE_file
WE_file = pd.read_excel('DataSet/ICNALE_Infosheet_201912.xlsx', sheet_name = 'Written Essays_2.4')
EE_file = pd.read_excel('DataSet/ICNALE_Infosheet_201912.xlsx', sheet_name = 'Edited Essays_2.1')
```

```python
EE_file.head()
```

```python
WE_file.dtypes
```

```python
EE_file.dtypes
```

```python
# Clean up column names - they're currently in the second row
EE_file.columns = EE_file.iloc[0]
EE_file = EE_file.drop(0)
```

```python
EE_file = EE_file.reset_index()
```

```python
EE_file.head()
```

```python
# Define and save the columns of interest - only the ID Code and the scores are of interest from EE_file: EE_scores
score_cols = ['Content (/12)', 'Organization (/12)', 'Vocabulary (/12)', 
              'Language Use (/12)', 'Mechanics (/12)', 'Total 1 (%)', 'Total 2 (Weighted %)']
EE_scores = EE_file[['File Name'] + score_cols].copy()
```

```python
EE_scores.head()
```

```python
# Rename ID to same identifier as WE_file
EE_scores = EE_scores.rename(columns = {'File Name' : 'Code'})
```

```python
# Format Code to be of same format as WE_file
for i, user in enumerate(EE_scores['Code']):
    EE_scores['Code'][i] = user[:6] + user[11:14]
```

```python
EE_scores.head()
```

```python
# How many unique Codes? Some are duplicates
EE_scores['Code'].nunique()
```

```python
# Convert scores to floats
EE_scores[score_cols] = EE_scores[score_cols].astype('float64', errors = 'raise')
```

```python
EE_scores.isna().sum()
```

```python
EE_scores.dtypes
```

```python
# Tidy EE_scores by averaging each individual's (Code's) essay scores
EE_unique = EE_scores.groupby('Code')[score_cols].mean()
```

```python
# Ready to merge
EE_unique.head()
```

```python
# Join EE_unique with WE_file. Left join with EE_unique - all entries without essay scores are meaningless
df = EE_unique.join(WE_file.set_index('Code'), how='left')
```

```python
df.shape
```

```python
df.dtypes
```

```python
# The following columns are the wrong format. Coerce to numeric type.
num_cols = ['Score', 'NS', 'Pronunciation', 'Presentation', 'EssayW']
```

```python
for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors = 'coerce')
```

```python
df.dtypes
```

```python
df = df.reset_index()
df.head()
```

```python
# View sample entry
df.iloc[0].T
```

```python
df.isna().sum()
```

```python
# Following columns are completely empty. Drop them.
df = df.drop(['PTJ2 (wds)', 'SMK2 (wds)', 'Self Ev', 'Yrs of Stay (< Yrs)', 'ENS Type'], axis=1).copy()
```

```python
# Following columns are not needed. Drop them.
df = df.drop(['Code', 'Content (/12)', 'Organization (/12)', 'Vocabulary (/12)',
       'Language Use (/12)', 'Mechanics (/12)', 'Total 2 (Weighted %)', 'INTM+INSM', 'INTM-INSM'], axis=1).copy()
```

```python
# Rename some awkward columns.
df = df.rename(columns = {
    'PTJ1 (wds)' : 'PTJ1', 
    'SMK1 (wds)' : 'SMK1',
    'Grade/Degree' : 'Grade', 
    'Major/ Occupation' : 'Major', 
    'Acad. Genre' : 'Field'})
```

```python
df.isna().sum()
```

```python
df.Sex.unique()
```

```python
# Unify to uppercase for Sex
df.Sex = df.Sex.str.upper()
```

```python
df.Grade.unique()
```

```python
# Deal with Grade
df['Grade_isna'] = df.Grade.isna().astype(int)
df.Grade = df.Grade.fillna(0)
```

```python
df.Major.unique()
```

```python
# Fill NaN Major/ Occupation to 'Other'
df.Major = df.Major.fillna('Other')
```

```python
df.Field.unique()
```

```python
# Fill NaN Acad. Genre to 'Other'
df.Field = df.Field.fillna('Other')
```

```python
# Check if the entry missing INTM is missing all the other single missing values as well
df[df.INTM.isna()].T
```

```python
# Fill the missing entries with median values instead of zeros and creating extra columns for all
missing_cols = ['INTM', 'INSM', 'Primary', 'Secondary', 'College', 'Inschool', 'Outschool', 
                'Listening', 'Reading', 'Speaking', 'Writing', 'NS', 'Pronunciation', 'Presentation', 'EssayW']
```

```python
for c in missing_cols:
    df[c] = df[c].fillna(df[c].median())
```

```python
# No more .isna() (except Test and Score - we'll deal with this later) - double check 
df.isna().sum()
```

```python
# Subset scores to their own columns
for key, group in df.groupby('Test'):
    df[key] = group['Score']
    print(key, ': ', df[key].nunique())
```

```python
# Not enough data in following tests - drop them
tests_to_drop = ['A Level (General Paper)', 'CSEPT', 'HKALE', 'IELTS', 'NMET', 
                 'O Level (Eng Lang)', 'TEPS', 'TOEFL (PBT)', 'TOEFL (iBT)']
df = df.drop(tests_to_drop, axis=1)
```

```python
# Fill the following tests with zeros for nan's
tests = ['CEE', 'ONET', 'TOEIC']

for test in tests:
    df[test + '_isna'] = df[test].isna().astype(int)
    df[test] = df[test].fillna(0)

df.head(20)
```

```python
# Test and Score columns are now redundant. Drop them.
df = df.drop(['Test', 'Score'], axis=1)
```
